```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import sklearn
```

```python
df=pd.read_csv("C:\\Users\\trigo\\Downloads\\Sales_April_2019.csv\\Sales_April_2019.csv")
df
```

```python
df.isnull().sum()
```

```python
df.dropna(inplace=True)
```

```python
df.isnull().sum()
```

```python
df
```

```python
df.info()
```

```python
non_numeric = df[~df["Price Each"].str.replace('.', '', 1).str.isnumeric()]
print(non_numeric)

# Remove non-numeric values
df = df[df["Price Each"].str.replace('.', '', 1).str.isnumeric()]

# Convert to numeric
df["Price Each"] = df["Price Each"].astype(float).astype(int)
```

```python
non_numeric = df[~df["Quantity Ordered"].str.replace('.', '', 1).str.isnumeric()]
print(non_numeric)

# Remove non-numeric values
df = df[df["Quantity Ordered"].str.replace('.', '', 1).str.isnumeric()]

# Convert to numeric
df["Quantity Ordered"] = df["Quantity Ordered"].astype(float).astype(int)
```

```python
df.info()
```

```python
df["Total_price"]=df["Price Each"]*df["Quantity Ordered"]
df["Total_price"]
```

```python
df
```

```python
df['Order Date'] = pd.to_datetime(df['Order Date'])
```

```python
df['Day'] = df['Order Date'].dt.day
```

```python
monthly_sales = df.groupby('Day')['Total_price'].sum()

import matplotlib.pyplot as plt

plt.figure(figsize=(20, 6))
monthly_sales.plot(kind='bar')
plt.title('Daily Sales')
plt.xlabel('Day')
plt.ylabel('Total Sales')
plt.xticks(rotation=0)
plt.show()
```
